### **Working with Object Lists**

Create a list called `cat_shelter` that stores 5 different `Cat` objects.

Then:
1. Loop through the list and call `.describe()` for each cat.
2. Write a function `oldest_cat(cat_list)` that returns the cat with the highest age.

```python
class Cat:
    def __init__(self, name="Unknown", age=0, color="Unknown"):
        self.name = name
        self.age = age
        self.color = color

    def meow(self):
        print(f"{self.name} says meow!")

    def describe(self):
        print(f"{self.name} is {self.age} years old and has {self.color} fur.")
cat_shelter = [
    Cat("Sirmeowmeow", 3, "black"),
    Cat("Curiosity", 5, "death"),
    Cat("whosagoodpuzzy", 2, "gray"),
    Cat("Garfield", 7, "Jerry is a horrible owner, who gives a cat lasagna?!?!"),
    Cat("Dog", 4, "Ran out of ideas at some point")
]
# you didn't say we couldn't have fun with it one of them is a color
for cat in cat_shelter:
    cat.describe()
def oldest_cat(cat_list):
    oldest = cat_list[0]
    for cat in cat_list:
        if cat.age > oldest.age:
            oldest = cat
    return oldest
oldest = oldest_cat(cat_shelter)
print(f"The oldest cat is {oldest.name}, who is {oldest.age} years old.")
```

### **Inheritance Challenge:**

Create a base class `Vehicle` with attributes:
1. `make`, `model`, `year`

Then create two **subclasses:**
1. `Car`: With additional attribute `num_doors`
2. `Truck`: With additional attribute `bed_length`

Each subclass should have a method:
1. `describe()`: that prints all attributes relevant to that vehicle type.

**Note**: In Python, subclasses **means** the same thing as a **child class**. In other programming languages, subclasses can be **confused** with **inner classes**, and **inner** classes are defined **INSIDE** of an **outer class**. When I **refer** to a subclass, I want you to create a **child class**, **NOT** an **inner class**.

```python
class Vehicle:
    def __init__(self, make="Unknown", model="Unknown", year=0):
        self.make = make
        self.model = model
        self.year = year


class Car(Vehicle):
    def __init__(self, make="Unknown", model="Unknown", year=0, num_doors=4):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def describe(self):
        print(f"{self.year} {self.make} {self.model} with {self.num_doors} doors.")


class Truck(Vehicle):
    def __init__(self, make="Unknown", model="Unknown", year=0, bed_length="Unknown"):
        super().__init__(make, model, year)
        self.bed_length = bed_length

    def describe(self):
        print(f"{self.year} {self.make} {self.model} with a {self.bed_length} bed.")


my_car = Car("Toyota", "Camry", 2022, 4)
my_truck = Truck("Ford", "F-150", 2021, "6-foot")

my_car.describe()
my_truck.describe()
```
