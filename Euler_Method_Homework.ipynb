## **What Is Euler’s Method?**

Euler’s Method approximates the solution of a differential equation of the form:

> **dy/dx = f(x, y)**

You use a small step size `h` to "walk forward" from an initial point `(x₀, y₀)` and estimate future values of `y`.

The update rule is:

> **yₙ₊₁ = yₙ + h * f(xₙ, yₙ)**  
> **xₙ₊₁ = xₙ + h**

---

## **Example**

Suppose you have this differential equation:

> **dy/dx = 2x**

That means the slope at any point `(x, y)` is `2x`.

Euler’s method would approximate the curve by repeating this:

```python
y = y + h * (2 * x)
x = x + h
```

With enough steps and a small enough `h`, this traces the curve.

---

## **Why This Is Useful**

We may **not know the original function y(x)**. But using Euler’s method, we can still **approximate** its shape using only the derivative!

- If `dy/dx = 2x` → `y = x²` (plus a constant). So if your derivative function is `2x`, Euler's method approximates your solution to be close to this function: `x²`.

- If `dy/dx = sin(x)` → `y = -cos(x)` (plus a constant)
- If `dy/dx = 1/x` → `y = ln(x)` (plus a constant)
